```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../data/daily_exchange_rates_USD.csv")
```

```python
data.drop([1, 2, 3, 4, 5, 6, 7], inplace=True)
```

```python
new_header = data.iloc[0]
data = data[1:]
data.columns = new_header
```

```python
data.rename(columns={"Reference area" : "date"}, inplace=True)
```

```python
data = data[data['date'] >= "1979-01-02"]
```

```python
data.reset_index(inplace=True)
```

```python
data.drop("index", axis=1, inplace=True)
```

```python
data = data.drop("level_0", axis=1)
```

```python
lista = list()
for column in data.columns:
    tmp = np.where(~data[column].isnull())
    lista.append([data.iloc[tmp[0][0]]['date'], column, tmp[0][0]])
sorted(lista)
```

```python
data.fillna(method="ffill", inplace=True)
data.fillna(method="bfill", inplace=True)
```
