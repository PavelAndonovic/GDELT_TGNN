```python
data_daily_exchange_rates_renamed = data_daily_exchange_rates.copy(deep=True)
```

```python
for col in data_daily_exchange_rates.columns[1:]:
    code = country_names_and_codes[country_names_and_codes['Country Name'] == col]['Country Code'].values
    if len(code) > 0:
        data_daily_exchange_rates_renamed.rename(columns={col: code[0]}, inplace=True)
```
