```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../data/yearly_HDI.csv", encoding = "latin-1")
```

```python
data.drop([col for col in data.columns if "Unnamed" in col], axis=1, inplace=True)
```

```python
data.drop("HDI Rank", axis=1, inplace=True)
```

```python
data = data.transpose()
```

```python
data_cols = data.iloc[0]
data = data[1:]
data.columns = data_cols
```

```python
data.index.names = ['date']
```

```python
data
```

```python
data.replace("..", np.nan, inplace=True)
```

```python
data.head(2)
```

```python
data.fillna(method="ffill", inplace=True)
data.fillna(method="bfill", inplace=True)
```

```python
data.isnull().sum()
```

```python
data.to_csv("../data/yearly_HDI.csv")
```
